# converting the model to onnx formate

| Part                                        | Meaning                                       |
| ------------------------------------------- | --------------------------------------------- |
| `paddle2onnx`                               | Runs the PaddlePaddle → ONNX conversion tool. |
| `--model_dir C:\Models\PP-OCRv5_server_rec` | Folder containing the Paddle model files.     |
| `--save_file C:\Models\rec.onnx`            | Where to save the converted ONNX model.       |
| `--opset_version 11`                        | ONNX operator set version to use.             |


```python
!pip install paddlepaddle
```

```python
!paddlex \
    --paddle2onnx \
    --paddle_model_dir /content/PP-OCRv5_server_det \
    --onnx_model_dir /content/det_onnx_model \
    --opset_version 11
```

```python
!paddlex \
    --paddle2onnx \
    --paddle_model_dir /content/PP-OCRv5_server_rec \
    --onnx_model_dir /content/rec_onnx_model \
    --opset_version 11
```
