# converting ocr output to label studio output 

```python
import json

def convert_ocr_to_ls(ocr_json, image_url, img_w, img_h):

    results = []

    polys = ocr_json["rec_polys"]
    texts = ocr_json["rec_texts"]

    for i, poly in enumerate(polys):

        xs = [p[0] for p in poly]
        ys = [p[1] for p in poly]

        x_min = min(xs)
        y_min = min(ys)

        x_max = max(xs)
        y_max = max(ys)

        w = x_max - x_min
        h = y_max - y_min

        results.append({
            "id": f"box_{i}",
            "type": "rectangle",
            "from_name": "bbox",
            "to_name": "image",
            "original_width": img_w,
            "original_height": img_h,
            "image_rotation": 0,
            "value": {
                "x": x_min/img_w*100,
                "y": y_min/img_h*100,
                "width": w/img_w*100,
                "height": h/img_h*100,
                "rotation": 0
            }
        })
        
        # OCR Text
        results.append({
            "id": f"box_{i}",
            "type": "textarea",
            "from_name": "transcription",
            "to_name": "image",
            "value": {
                "text": [texts[i]]
            }
        })
        
    return {
        "data": {
            "ocr": image_url
        },
        "predictions":[
            {
                "model_version":"paddleocr",
                "result": results
            }
        ]
    }

data=convert_ocr_to_ls(res, "http://localhost:8000/output/1.jpg", 2232, 3300)
print(data)
                       
# with open("./output/paddle_ocr_result.json", "r") as f:
#     data = json.load(f)
#     print(data.get('ocr'))
with open("output/labelstudio_tasks.json", "w") as f:
    json.dump(data, f, indent=4)
```

```python
import json


# Load your Label Studio JSON
with open("output/labelstudio_output.json", 'r') as f:
    data = json.load(f)

    

def normalize_bbox(x, y, width, height):
    """
    Convert % bbox → LayoutLM format (0–1000)
    """
    x1 = int(x * 10)
    y1 = int(y * 10)
    x2 = int((x + width) * 10)
    y2 = int((y + height) * 10)
    print("bbox")
    return [x1, y1, x2, y2]
    


def convert_label(label):
    """
    Convert labels to BIO format
    """
    print("label")
    if label.lower() == "ignore":
        return "O"
    return label.upper()

```

```python

final_dataset = []

for task in data:
    print(1)
    print(task["label"][0]["labels"][0])
    annotations = task["label"][0]["labels"][0]
    if not annotations:
        continue
    print(annotations)
    results = annotations["label"]

    words = []
    bboxes = []
    labels = []
    print(3)



    
    for item in results:
        print(4)
        if item["type"] != "labels":
            continue

        value = item["value"]

        # Extract data
        text = value.get("text", "")
        x = value.get("x", 0)
        y = value.get("y", 0)
        width = value.get("width", 0)
        height = value.get("height", 0)
        label = value.get("labels", ["O"])[0]

        # Split words
        tokens = text.split()

        # Convert bbox
        bbox = normalize_bbox(x, y, width, height)

        # Assign BIO labels
        for i, token in enumerate(tokens):
            words.append(token)
            bboxes.append(bbox)

            if label.lower() == "ignore":
                labels.append("O")
            else:
                if i == 0:
                    labels.append(f"B-{label.upper()}")
                else:
                    labels.append(f"I-{label.upper()}")

    final_dataset.append({
        "words": words,
        "bboxes": bboxes,
        "labels": labels
    })


```

```python
# Save cleaned dataset
with open("layoutlm_dataset.json", "w") as f:
    json.dump(final_dataset, f, indent=4)


print("✅ Dataset ready for LayoutLMv3 training!")
```

```python
import torch
from transformers import LayoutLMv3Tokenizer, LayoutLMv3ForTokenClassification
from paddleocr import PaddleOCR
from PIL import Image
from torchvision import transforms
```

```python

# ---------------------------
# 1. Setup
# ---------------------------
device = "cuda" if torch.cuda.is_available() else "cpu"

tokenizer = LayoutLMv3Tokenizer.from_pretrained("microsoft/layoutlmv3-base")
model = LayoutLMv3ForTokenClassification.from_pretrained("microsoft/layoutlmv3-base")

import torch
print(torch.cuda.is_available())  # Should print True
print(torch.cuda.get_device_name(0))  # Shows GPU name

model.to(device)
model.eval()

ocr = PaddleOCR(use_angle_cls=True, lang='en')  # GPU if configured
```

```python

# ---------------------------
# 2. Load image
# ---------------------------
image_path = "./training data/1.jpg"
image = Image.open(image_path).convert("RGB")
width, height = image.size

print(width,height)
```

```python

# ---------------------------
# 3. Run PaddleOCR
# ---------------------------
ocr_result = ocr.predict(image_path)

words = []
boxes = []

for line in ocr_result[0]:
    box, text, conf = line

    words.append(text)

    x_coords = [p[0] for p in box]
    y_coords = [p[1] for p in box]

    x0, x1 = min(x_coords), max(x_coords)
    y0, y1 = min(y_coords), max(y_coords)

    boxes.append([x0, y0, x1, y1])

pprint.pprint(boxes)
```

```python

# ---------------------------
# 4. Normalize boxes (0-1000)
# ---------------------------
def normalize_box(box, width, height):
    return [
        int(1000 * box[0] / width),
        int(1000 * box[1] / height),
        int(1000 * box[2] / width),
        int(1000 * box[3] / height),
    ]

normalized_boxes = [normalize_box(b, width, height) for b in boxes]
```

```python

# ---------------------------
# 5. Tokenize
# ---------------------------
encoding = tokenizer(
    words,
    padding="max_length",
    truncation=True,
    return_tensors="pt",
    is_split_into_words=True
)
```

```python

# ---------------------------
# 6. Align boxes with tokens
# ---------------------------
word_ids = encoding.word_ids()


bbox = []
for word_idx in word_ids:
    if word_idx is None:
        bbox.append([0, 0, 0, 0])
    else:
        bbox.append(normalized_boxes[word_idx])

encoding["bbox"] = torch.tensor([bbox])

# ---------------------------
# 7. Image → pixel_values
# ---------------------------
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
])

pixel_values = transform(image).unsqueeze(0)
encoding["pixel_values"] = pixel_values

# ---------------------------
# 8. Move to GPU
# ---------------------------
encoding = {k: v.to(device) for k, v in encoding.items()}

# ---------------------------
# 9. Inference
# ---------------------------
with torch.no_grad():
    outputs = model(**encoding)

logits = outputs.logits
predictions = torch.argmax(logits, dim=-1)

# ---------------------------
# 10. Decode output
# ---------------------------
tokens = tokenizer.convert_ids_to_tokens(encoding["input_ids"][0])

for token, pred in zip(tokens, predictions[0]):
    print(f"{token:15} -> {pred.item()}")
```
