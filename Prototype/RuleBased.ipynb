```python
import cv2
import numpy as np
image_path1="./training data/diff pattren img/2.jpg"
img = cv2.imread(image_path1)

if img is None:
    print("Image not loaded. Check path!")

# # Convert to binary (important!)
# _, binary = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

# # Create kernel (controls thickness)
# kernel = np.ones((5, 5), np.uint8)

# # Apply dilation
# thick = cv2.dilate(binary, kernel, iterations=1)

#----------------------------------
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# largest_contour = max(contours, key=cv2.contourArea)
# cv2.drawContours(img, largest_contour, -1, (0,255,0), 4)
# cv2.imwrite('./output/opencv/img.jpeg', img)
# print(largest_contour)
# contours, _ = cv2.findContours(gray, cv2.RETR_REC, cv2.CHAIN_APPROX_SIMPLE)
# cv2.drawContours(gray, contours, -1, (0,255,0), 4)

cv2.imwrite('./output/opencv/gray1.jpeg', gray)


clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
gray = clahe.apply(gray)

cv2.imwrite('./output/opencv/gray2.jpeg', gray)

# ✅ Better threshold (adaptive handles uneven lighting)
thresh = cv2.adaptiveThreshold(
    gray, 255,
    cv2.ADAPTIVE_THRESH_MEAN_C,
    cv2.THRESH_BINARY_INV,
    15, 2
)

kernel = np.zeros((3,3), np.uint8)
thresh = cv2.erode(thresh, kernel, iterations=4)

cv2.imwrite('./output/opencv/thresh1.jpeg', thresh)


# --- Horizontal lines ---
horizontal_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (60,1))
horizontal = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, horizontal_kernel)
cv2.imwrite('./output/opencv/horizontal.jpeg', horizontal)
# 🔥 connect broken horizontal lines
horizontal = cv2.dilate(horizontal, np.ones((3,3),np.uint8), iterations=2)
cv2.imwrite('./output/opencv/horizontal1.jpeg', horizontal)
# --- Vertical lines ---
vertical_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (1,60))
vertical = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, vertical_kernel)
cv2.imwrite('./output/opencv/vertical.jpeg', vertical)
# 🔥 connect broken vertical lines
vertical = cv2.dilate(vertical, np.ones((3,3),np.uint8), iterations=2)
cv2.imwrite('./output/opencv/vertical1.jpeg', vertical)
# Combine
boxes = cv2.add(horizontal, vertical)
cv2.imwrite('./output/opencv/boxes.jpeg', boxes)
# 🔥 Final closing to join gaps
kernel = np.ones((10,10), np.uint8)
boxes = cv2.morphologyEx(boxes, cv2.MORPH_CLOSE, kernel, iterations=2)
cv2.imwrite('./output/opencv/boxes1.jpeg', boxes)




# kernel = np.ones((3,6), np.uint8)
# dilated = cv2.dilate(boxes, kernel, iterations=2)
# cv2.imwrite('./output/opencv/boxes4.jpeg', dilated)
#make the img border small
# Define a kernel (3x3)
kernel = np.ones((3,3), np.uint8)
# Apply erosion
eroded = cv2.erode(boxes, kernel, iterations=1)
cv2.imwrite('./output/opencv/boxes2.jpeg', eroded)

# Find contours
contours, _ = cv2.findContours(boxes, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
# cv2.imwrite('./output/opencv/contours.jpeg', contours)

output = img.copy()

image_out = cv2.imread('./output/opencv/boxes1.jpeg')

cv2.drawContours(output, contours, -1, (0,255,0), 1)
# for cnt in contours:
    # x,y,w,h = cv2.boundingRect(cnt)
    # cv2.line(image_out, cnt[0], cnt[1], (0, 255, 0), 2)
    # filter noise
    # if w > 40 and h > 20:
    # cv2.rectangle(image_out, (x,y), (x+w,y+h), (0,255,255), 2)
cv2.imwrite('./output/opencv/output.jpeg', output)
# cv2.imshow("Fixed Boxes", output)
# cv2.waitKey(0)
```
